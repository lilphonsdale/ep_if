```python
import pandas as pd
```

```python
#reading the test results data from the cleaned excel file

results = "ELA Results.csv"
results_df = pd.DataFrame(pd.read_csv(results))
new_header = results_df.iloc[0]
results_df = results_df[1:]
results_df.columns = new_header
results_df.head()
```

```python
#creating a list of the schools in the pilot program

pilot_schools = [
    "610534",
    "609880",
    "610066",
    "610082",
    "610182" 
]
```

```python
#filtering the dataset to show only the pilot schools' 5 graders in 2017

pilot_participants = results_df[results_df['School ID'].isin(pilot_schools)]
pilot_participants = pilot_participants[results_df['Test Name']=="ELA/Literacy Grade 5"]
pilot_participants_2017 = pilot_participants[results_df['Year']=="2017"]
pilot_participants_2017
```

```python
#getting summary statistics for the pilot schools 2017 5 grade tests

pilot_participants_2017["Overall ELA Mean Scale Score"].astype(int).describe()
```

```python
#filtering the dataset to show only the pilot schools' 5 graders in 2016

pilot_participants = results_df[results_df['School ID'].isin(pilot_schools)]
pilot_participants = pilot_participants[results_df['Test Name']=="ELA/Literacy Grade 5"]
pilot_participants_2016 = pilot_participants[results_df['Year']=="2016"]
pilot_participants_2016
```

```python
#filtering the dataset to show only the non-pilot schools' 5 graders in 2017

non_pilot = results_df[results_df['School ID'].isin(pilot_schools) == False]
non_pilot = non_pilot[results_df['Test Name']=="ELA/Literacy Grade 5"]
non_pilot_2017 = non_pilot[results_df['Year']=="2017"]
non_pilot_2017
```

```python
#getting summary statistics for the non pilot schools

non_pilot_2017["Overall ELA Mean Scale Score"].astype(int).describe()
```

```python
# creating a new dataframe to compare results from year to year

growth = pd.merge(left=pilot_participants_2017, right=pilot_participants_2016,how='right',on='School ID')
growth = growth[["School ID","Overall ELA Mean Scale Score_x","Overall ELA Mean Scale Score_y"]]
growth
```

```python
#changing the data type to be able to perform calculations

growth = growth.astype(int)
growth.dtypes
```

```python
#calculating change in test scores from year to year

for i in growth:
    growth["change"] = growth["Overall ELA Mean Scale Score_x"] - growth["Overall ELA Mean Scale Score_y"]
    growth["percent_change"] = ((growth["Overall ELA Mean Scale Score_x"] - growth["Overall ELA Mean Scale Score_y"])/growth["Overall ELA Mean Scale Score_y"])*100

growth.sort_values(by="percent_change", ascending=False)
```
